```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
import string 
```

```python
def make_soup(url):
    html = requests.get(url)
    plain = html.text
    soup = BeautifulSoup(plain, 'html.parser')
    return soup

def crawler(seed):
    
    Q = [seed]
    brand = []
    model = []
    upfront_cost = [] 
    monthly_cost = [] 
    
    for page in range (0,5):
        if page == 0: 
            page_no =  1
        else:
            page_url = seed +'?page='+str(page)
            Q.append(page_url)
            
    while(Q!=[]):
        next_url = Q[0]
        Q = Q[1:]
        
        print('CRAWLER IS CURRENTLY ON THIS LINK', next_url)
        page_soup = make_soup(next_url)
        O2_publications = page_soup.find_all(attrs={"component-name" : "productTile"})
        
        #print(len(O2_publications))
        
        for pub in O2_publications:
            
           # print(pub.find("div",{'class':'details'}).text)
            #print(pub.find("div",{'class':'upfront-cost-details'}).text)
            #print(pub.find("div",{'class':'monthly-cost-details'}).text)
            #print("-------------------------------")
            
        
           
            #find Brand
            pub_brand = pub.find("span",{'class':'brand-name col-xs-12'}).text
            brand.append(pub_brand)
            print('--------------------------------------------------------------------------------------------------')
            print("Brand Manufacturer: ", pub_brand)
            
            
            #find model
            pub_model = pub.find('div', {'class':'model-family col-xs-12'}).text
            model.append(pub_model)
            print("Manufacturer Model: ", pub_model)
            
            #find upfront cost
            pub_upfront_cost = pub.find("span",{'class':'headline'}).text
            upfront_cost.append(pub_upfront_cost)
            print("Upfront Cost: ", pub_upfront_cost)
            
            #find monthly cost
            pub_monthly_cost = pub.find("span",{'class':['costVal','costPence']}).text
            monthly_cost.append(pub_monthly_cost)
            print("Monthly Cost: ", pub_monthly_cost)
            print('-------------------------------------------------------------------------------------------------')
            
   

    Results = {'Brand'                 : brand,
               'Model'                 : model,
               'Upfront_cost'          : upfront_cost,
               'Monthly_Cost'          : monthly_cost
              }

    df_results = pd.DataFrame(Results)
   # df_results.to_csv('C:/Users/Abrar Ahmad/Desktop/O2_data.csv',index=False,encoding='utf-8')

    return(brand,model,upfront_cost,monthly_cost,df_results)
```

```python
crawler("https://www.o2.co.uk/shop/phones")
```
